## Imports

```python
import pandas as pd
import sqlite3
```

## create a connection to the database using the library sqlite3

```python
connect = sqlite3.connect('../data/checking-logs.sqlite.sqlite_copy')
```

## get the schema of the table checker

```python
request = 'PRAGMA table_info(checker);'
pd.io.sql.read_sql(request, connect)
```

## get only the first 10 rows of the table checker to check what the table looks like

```python
request = 'SELECT * FROM checker LIMIT 10;'
pd.io.sql.read_sql(request, connect)
```

## count how many rows satisfy the following conditions using only one query with any number of subqueries:

```python
request = '''
SELECT COUNT(*)
FROM pageviews
WHERE pageviews.uid IN
    (SELECT checker.uid
    FROM checker
    WHERE status='ready'
    AND numTrials=1
    AND labname in ('laba04', 'laba04s', 'laba05', 'laba06', 'laba06s', 'project1')
    )
'''
checkers = pd.io.sql.read_sql(request, connect)
```

```python
checkers.columns = ['cnt']
checkers
```

## close the connection

```python
connect.close()
```
