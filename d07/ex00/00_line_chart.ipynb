## Imports

```python
import pandas as pd
import sqlite3
```

## make a connection to the database (it is the same as the previous day)

```python
connect = sqlite3.connect('../data/checking-logs.sqlite.sqlite')
```

## run a query that gets the datetime from the pageviews table, selecting only the users and not the admins

```python
request = '''
SELECT *
FROM pageviews
WHERE uid LIKE 'user_%'
'''
df = pd.io.sql.read_sql(request, connect, parse_dates=['datetime'])
df
```

## using Pandas, create a new dataframe where the visits are counted and grouped by date

```python
new_df = df.groupby(by=df['datetime'].dt.date).count()['uid'].reset_index()
new_df.columns = ['date', 'views']
new_df = new_df.set_index('date')
new_df
```

## using Pandas method .plot(), create a graph

```python
new_df.plot(fontsize=8,
            figsize=(15,8),
            title='Views per day',
            rot=90)
```

## close the connection to the database

```python
connect.close()
```
