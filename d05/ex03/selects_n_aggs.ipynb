## load the JSON file that you created in the previous exercise into a dataframe

```python
df = pd.read_json('../data/auto.json', orient='records').set_index('CarNumber')
df.head()
```

```python
df.info()
```

## make the following selects

## make the aggregations with the make and the model

```python
df.groupby(by=['Make']).median()['Fines']
```

```python
df.groupby(by=['Make', 'Model']).median()['Fines']
```

```python
df.groupby(by=['Make', 'Model']).count()['Fines']
```

```python
df.groupby(['Make', 'Model']).min()['Fines']
```

```python
df.groupby(['Make', 'Model']).max()['Fines']
```

```python
df.groupby(['Make', 'Model']).std()['Fines']
```

## make the aggregations with the car number

```python
count_groupped = df.groupby(['CarNumber']).count()['Fines'].reset_index(
).sort_values('Fines', ascending=False).set_index('CarNumber')
count_groupped
```

```python
df.loc[count_groupped.head(1).index[0]]
```

```python
sum_groupped = df.groupby(['CarNumber']).sum()['Fines'].reset_index(
).sort_values('Fines', ascending=False).set_index('CarNumber')
sum_groupped
```

```python
df.loc[sum_groupped.head(1).index[0]]
```

```python
df.groupby(['CarNumber', 'Make']).count().sum()
```

```python
new_df = df[['Make', 'Model']].reset_index().drop_duplicates().groupby(
    ['CarNumber'])['Make'].count().reset_index().sort_values('Make', ascending=False)
df.loc[new_df[new_df['Make'] > 1]['CarNumber']]
```
